```python

import pandas as pd
import numpy as np
import holidays
from pathlib import Path

RAW_PATH = "thesis-data.csv"
OUT_PATH = "thesis-data-featured.csv"

# 1) Load raw
df = pd.read_csv(RAW_PATH)

# 2) Parse & validate DATETIME
if "DATETIME" not in df.columns:
    raise KeyError(f"'DATETIME' not found. Columns: {list(df.columns)}")

df["DATETIME"] = pd.to_datetime(df["DATETIME"], errors="coerce")
df = df.dropna(subset=["DATETIME"]).sort_values("DATETIME").reset_index(drop=True)

# 3) Pick/validate target column
target = "ENGLAND_WALES_DEMAND"
if target not in df.columns:
    guess = next((c for c in df.columns if "demand" in c.lower()), None)
    if guess is None:
        raise KeyError(f"'{target}' not found and no column contains 'demand'. Columns: {list(df.columns)}")
    target = guess

# 4) Time features
df["day_of_week"] = df["DATETIME"].dt.dayofweek      # 0=Mon
df["month"]       = df["DATETIME"].dt.month
df["hour"]        = df["DATETIME"].dt.hour
df["minute"]      = df["DATETIME"].dt.minute
df["is_weekend"]  = (df["day_of_week"] >= 5).astype(int)
df["settlement_period"] = df["hour"] * 2 + (df["minute"] // 30) + 1  # 1..48

# 5) Lags & rolling (needs sorted datetime)
df["lag_1"]  = df[target].shift(1)
df["lag_48"] = df[target].shift(48)
df["rolling_mean_7d"] = df[target].rolling(window=48*7, min_periods=48*7).mean()

# 6) Renewable ratios (safe division; keep NaN if source missing)
for src, new in [
    ("EMBEDDED_WIND_GENERATION",  "wind_ratio"),
    ("EMBEDDED_SOLAR_GENERATION", "solar_ratio"),
]:
    if src in df.columns:
        denom = df[target].replace({0: np.nan})
        df[new] = df[src] / denom
    else:
        df[new] = np.nan

# 7) UK holidays
years = range(int(df["DATETIME"].dt.year.min()), int(df["DATETIME"].dt.year.max()) + 1)
uk_holidays = holidays.UnitedKingdom(years=years)
df["is_holiday"] = df["DATETIME"].dt.date.astype(str).isin([str(d) for d in uk_holidays]).astype(int)

# 8) Final clean: drop rows where lag/rolling are NaN (head of series)
df = df.dropna(subset=["lag_1", "lag_48", "rolling_mean_7d"]).reset_index(drop=True)

# 9) Save
df.to_csv(OUT_PATH, index=False)

# 10) Quick check
print(" Feature engineering complete.")
print(f"Saved to: {Path(OUT_PATH).resolve()}")
show_cols = [c for c in [target, "lag_1", "lag_48", "rolling_mean_7d", "wind_ratio", "solar_ratio", "is_holiday"] if c in df.columns]
print(df[ ['DATETIME'] + show_cols ].head())
print("\nColumns:", list(df.columns))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

 
df = pd.read_csv('thesis-data-featured.csv', parse_dates=['DATETIME'])

# Rename columns

df = df.rename(columns={'DATETIME': 'Date', 'ENGLAND_WALES_DEMAND': 'Demand'})
df.set_index('Date', inplace=True)

# Keep only target column, drop missing values

df = df[['Demand']].dropna()

# Train-Test Split (80% train, 20% test)

split_index = int(len(df) * 0.8)
train = df.iloc[:split_index]
test = df.iloc[split_index:]

#  Fit ARIMA model (order can be tuned)

model = ARIMA(train['Demand'], order=(5, 1, 0))
model_fit = model.fit()

# Forecast

forecast = model_fit.forecast(steps=len(test))
forecast.index = test.index  # Fix: align forecast index with test index

# Evaluation Metrics
mae = mean_absolute_error(test['Demand'], forecast)
rmse = np.sqrt(mean_squared_error(test['Demand'], forecast))
mape = np.mean(np.abs((test['Demand'] - forecast) / test['Demand'])) * 100
accuracy = 100 - mape

#  Print Results
print(" ARIMA Model Performance:")
print("MAE:", round(mae, 2))
print("RMSE:", round(rmse, 2))
print("MAPE:", round(mape, 2), "%")
print("Accuracy:", round(accuracy, 2), "%")

# Plot Forecast vs Actual
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['Demand'], label='Train')
plt.plot(test.index, test['Demand'], label='Actual')
plt.plot(test.index, forecast, label='ARIMA Forecast')
plt.title('ARIMA Forecast vs Actual Electricity Demand')
plt.xlabel('Date')
plt.ylabel('Electricity Demand (MW)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

 
df = pd.read_csv('thesis-data-featured.csv', parse_dates=['DATETIME'])

#  Rename and format 

df = df.rename(columns={'DATETIME': 'Date', 'ENGLAND_WALES_DEMAND': 'Demand'})
df.set_index('Date', inplace=True)
df = df.dropna()

# Select features (no lag_1 or lag_48)

features = ['rolling_mean_7d', 'wind_ratio', 'solar_ratio',
            'hour', 'day_of_week', 'is_weekend']

X = df[features]
y = df['Demand']

# Train-Test split (time-based 80/20)

split_index = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

# Feature Scaling (important for SVR)

scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1,1)).ravel()

# Train SVR model (RBF kernel is good default)

svr = SVR(kernel='rbf', C=100, epsilon=0.1)
svr.fit(X_train_scaled, y_train_scaled)

# Predict (and inverse scale)
y_pred_scaled = svr.predict(X_test_scaled)
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1,1)).ravel()

# Evaluation

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
accuracy = 100 - mape

# Print results

print(" SVR Model Performance:")
print("MAE:", round(mae, 2))
print("RMSE:", round(rmse, 2))
print("MAPE:", round(mape, 2), "%")
print("Accuracy:", round(accuracy, 2), "%")

# Plot forecast vs actual

plt.figure(figsize=(12,6))
plt.plot(y_test.index, y_test.values, label='Actual')
plt.plot(y_test.index, y_pred, label='Predicted (SVR)', alpha=0.7)
plt.title('SVR Forecast vs Actual')
plt.xlabel('Date')
plt.ylabel('Electricity Demand (MW)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

 
df = pd.read_csv('thesis-data-featured.csv', parse_dates=['DATETIME'])
df = df.rename(columns={'DATETIME': 'Date', 'ENGLAND_WALES_DEMAND': 'Demand'})
df.set_index('Date', inplace=True)
df = df[['Demand']].dropna()

# Scale demand between 0 and 1

scaler = MinMaxScaler()
demand_scaled = scaler.fit_transform(df)

# Create time sequences

def create_sequences(data, window_size=24):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

# Use 24 previous time steps to predict the next (can tune this)

window_size = 24
X, y = create_sequences(demand_scaled, window_size)

#  Train-test split (80-20)

split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

#  Reshape input for LSTM: [samples, time steps, features]

X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Build the LSTM model

model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(window_size, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

#  Train the model

model.fit(X_train, y_train, epochs=20, verbose=1)

# Predict and inverse scale

y_pred_scaled = model.predict(X_test)
y_pred = scaler.inverse_transform(y_pred_scaled)
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# Evaluate the model

mae = mean_absolute_error(y_test_actual, y_pred)
rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred))
mape = np.mean(np.abs((y_test_actual - y_pred) / y_test_actual)) * 100
accuracy = 100 - mape

print(" LSTM Model Performance:")
print("MAE:", round(mae, 2))
print("RMSE:", round(rmse, 2))
print("MAPE:", round(mape, 2), "%")
print("Accuracy:", round(accuracy, 2), "%")

# Plot actual vs predicted
plt.figure(figsize=(12,6))
plt.plot(y_test_actual, label='Actual')
plt.plot(y_pred, label='Predicted (LSTM)', alpha=0.7)
plt.title('LSTM Forecast vs Actual')
plt.xlabel('Time Steps')
plt.ylabel('Electricity Demand (MW)')
plt.legend()
plt.tight_layout()
plt.show()

```

```python
#  Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

 
df = pd.read_csv('thesis-data-featured.csv', parse_dates=['DATETIME'])
df = df.rename(columns={'DATETIME': 'Date', 'ENGLAND_WALES_DEMAND': 'Demand'})
df.set_index('Date', inplace=True)
df = df.dropna()

# Features for RF (excluding lag_1 and lag_48 for realism)

rf_features = ['rolling_mean_7d', 'wind_ratio', 'solar_ratio',
               'hour', 'day_of_week', 'is_weekend']

X = df[rf_features]
y = df['Demand']

# Train-test split

split_index = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

# Train Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

# Compute Residuals (actual - RF prediction)
residuals = y_test.values - rf_pred

# Prepare residuals for LSTM

# We'll train LSTM to predict residuals based on past residual values

scaler = MinMaxScaler()
res_scaled = scaler.fit_transform(residuals.reshape(-1, 1))

def create_sequences(data, window_size=24):
    X_seq, y_seq = [], []
    for i in range(window_size, len(data)):
        X_seq.append(data[i-window_size:i, 0])
        y_seq.append(data[i, 0])
    return np.array(X_seq), np.array(y_seq)

X_lstm, y_lstm = create_sequences(res_scaled, window_size=24)

# Reshape for LSTM

X_lstm = X_lstm.reshape((X_lstm.shape[0], X_lstm.shape[1], 1))

#  Match lengths for final prediction

rf_pred_trimmed = rf_pred[24:]
y_test_trimmed = y_test.values[24:]

# Build and train LSTM

model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(X_lstm.shape[1], 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X_lstm, y_lstm, epochs=20, verbose=1)

#  Predict LSTM residuals and add to RF predictions

residual_preds = model.predict(X_lstm)
residual_preds_inverse = scaler.inverse_transform(residual_preds)
final_preds = rf_pred_trimmed + residual_preds_inverse.flatten()

#  Evaluation
mae = mean_absolute_error(y_test_trimmed, final_preds)
rmse = np.sqrt(mean_squared_error(y_test_trimmed, final_preds))
mape = np.mean(np.abs((y_test_trimmed - final_preds) / y_test_trimmed)) * 100
accuracy = 100 - mape

print(" Hybrid RF + LSTM Model Performance:")
print("MAE:", round(mae, 2))
print("RMSE:", round(rmse, 2))
print("MAPE:", round(mape, 2), "%")
print("Accuracy:", round(accuracy, 2), "%")

#  Plotting
plt.figure(figsize=(12, 6))
plt.plot(y_test_trimmed, label='Actual')
plt.plot(final_preds, label='Hybrid RF+LSTM Prediction', alpha=0.7)
plt.title('Hybrid Model: Random Forest + LSTM')
plt.xlabel('Time Steps')
plt.ylabel('Electricity Demand (MW)')
plt.legend()
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

#  Load dataset
df = pd.read_csv('thesis-data-featured.csv', parse_dates=['DATETIME'])
df = df.rename(columns={'DATETIME': 'Date', 'ENGLAND_WALES_DEMAND': 'Demand'})
df.set_index('Date', inplace=True)
df = df.dropna()

#  Features for MLP (with lags)
features = ['lag_1', 'lag_48', 'rolling_mean_7d', 'wind_ratio', 'solar_ratio',
            'hour', 'day_of_week', 'is_weekend']

X = df[features]
y = df['Demand']

#  Train-Test Split
split_index = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

#  Scale features
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).ravel()

#  Train MLPRegressor
mlp = MLPRegressor(hidden_layer_sizes=(64, 32), activation='relu', solver='adam',
                   max_iter=500, random_state=42)
mlp.fit(X_train_scaled, y_train_scaled)

#  Predict and inverse transform
y_pred_scaled = mlp.predict(X_test_scaled)
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).ravel()

#  Evaluate performance
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
accuracy = 100 - mape

print(" Scikit-learn MLP Model Performance:")
print("MAE:", round(mae, 2))
print("RMSE:", round(rmse, 2))
print("MAPE:", round(mape, 2), "%")
print("Accuracy:", round(accuracy, 2), "%")

#  Plot Actual vs Predicted
plt.figure(figsize=(12,6))
plt.plot(y_test.index, y_test.values, label='Actual')
plt.plot(y_test.index, y_pred, label='Predicted (MLP)', alpha=0.7)
plt.title('MLP Forecast vs Actual (Scikit-learn)')
plt.xlabel('Date')
plt.ylabel('Electricity Demand (MW)')
plt.legend()
plt.tight_layout()
plt.show()
```
